## Preparación de los datos

**Se aplica normalización y con estandarización al dataset para tener todas las variables independientes en una misma escala**

```python
sc = StandardScaler()
#Para el modelo 2
X_train_1 = sc.fit_transform(X_train_1)
X_test_1 = sc.transform(X_test_1)
#Para el modelo 3
X_train_2 = sc.fit_transform(X_train_2)
X_test_2 = sc.transform(X_test_2)
```

## Modelación Lineal

**Modelo 1 - Sin utilizar ninguna librería**

```python
x_model = data['bmi'].to_numpy()
y_model = data['charges'].to_numpy()


x_mean = np.mean(x_model)
y_mean = np.mean(y_model)

n = 0
d = 0
for i in range(len(x_model)):
    n += (x_model[i] - x_mean) * (y_model[i] - y_mean)
    d += (x_model[i] - x_mean) ** 2
m = n / d
c = y_mean - (m * x_mean)
 
max_x =  + 100
min_x =  - 100
 
# Calculating line values x and y
x_line = np.linspace(np.min(x_model), np.max(x_model), len(x_model))
y_line = (m * x_line) + c
 
plt.plot(x_line, y_line, label='Regression Line')
plt.scatter(x_model, y_model, c='red', label='Data')
plt.legend()
plt.show()
```

## Evaluación de los modelos

**Modelo 1:**

Se calcula el Root Mean Squares Error:

```python
rmse = 0
for i in range(len(x_model)):
    y_pred_model1 = c + m * x_model[i]
    rmse += (y_model[i] - y_pred_model1) ** 2
rmse = np.sqrt(rmse/n)
print(rmse)
```

Se calcula $R^2$:

```python
total = 0
rest = 0
for i in range(len(x_model)):
    y_pred_model1 = c + m * x_model[i]
    total += (y_model[i] - y_mean) ** 2
    rest += (y_model[i] - y_pred_model1) ** 2
r2 = 1 - (rest/total)
print(r2)
```

## Regresión Polinomial

```python
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt
```

**Modelo 2:**

```python
model = make_pipeline(PolynomialFeatures(2), Ridge())
model.fit(X_train_1, y_train_1)
y_plot = model.predict(X_test_1)
plt.plot(X_test_1, y_plot, color='red', linewidth=2, label="Regresión polinomial de grado 2")

plt.legend()

plt.show()
```

**Modelo 2:**

```python
model = make_pipeline(PolynomialFeatures(2), Ridge())
model.fit(X_train_2, y_train_2)
y_plot = model.predict(X_test_2)
plt.plot(X_test_2, y_plot, color='blue', linewidth=2, label="Regresión polinomial de grado 2")

plt.legend()

plt.show()
```
